```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import mean_squared_error
import xgboost as xgb
from xgboost import plot_importance
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
input_train=pd.read_csv("train.csv",encoding="gb2312")
input_test=pd.read_csv("test.csv",encoding="gb2312")
example=pd.read_csv("submit_example.csv",encoding="gb2312")
```

```python
df1_train=input_train.iloc[1:,[x for x in range(23)]+[23]]
df2_train=input_train.iloc[1:,[x for x in range(23)]+[24]]
df3_train=input_train.iloc[1:,[x for x in range(23)]+[25]]
df4_train=input_train.iloc[1:,[x for x in range(23)]+[26]]
```

```python
poly = PolynomialFeatures(degree=3)
np_train = poly.fit_transform(df1_train.iloc[:,[1,2,3,4,5,8]])
df_new_features=pd.DataFrame(np_train)
df_new_features.head()
```

```python
df_train = pd.concat([pd.DataFrame(df1_train.iloc[:,:23].values),df_new_features.iloc[:,7:]],axis=1)
df_train.head()
```

```python
df_test_value=input_test.iloc[1:,:]
np_test = poly.fit_transform(df_test_value.iloc[:,[1,2,3,4,5,8]])
df_test = pd.concat([pd.DataFrame(df_test_value.iloc[:,:].values),pd.DataFrame(np_test).iloc[:,7:]],axis=1)
df_test.head()
```

```python
# model1 = xgb.XGBRegressor(n_estimators=1100,n_jobs=-1,max_depth=5,colsample_bytree=0.2,subsample=0.7)
# model2 = xgb.XGBRegressor(n_estimators=1100,n_jobs=-1,max_depth=5,colsample_bytree=0.2,subsample=0.7)
# model3 = xgb.XGBRegressor(n_estimators=1100,n_jobs=-1,max_depth=5,colsample_bytree=0.2,subsample=0.7)
# model4 = xgb.XGBRegressor(n_estimators=1100,n_jobs=-1,max_depth=5,colsample_bytree=0.2,subsample=0.7)
model1 = xgb.XGBRegressor(learning_rate=0.05,n_estimators=1100,n_jobs=-1,max_depth=3,colsample_bytree=0.2,subsample=0.7,min_child_weight=2)
model2 = xgb.XGBRegressor(learning_rate=0.31,n_estimators=1100,n_jobs=-1,max_depth=7,colsample_bytree=0.2,subsample=1,min_child_weight=2.5)
model3 = xgb.XGBRegressor(learning_rate=0.15,n_estimators=1100,n_jobs=-1,max_depth=2,colsample_bytree=0.1,subsample=0.6,min_child_weight=1.5)
model4 = xgb.XGBRegressor(learning_rate=0.1,n_estimators=1100,n_jobs=-1,max_depth=7,colsample_bytree=0.2,subsample=0.7,min_child_weight=1)
```

```python
model1.fit(df_train.iloc[:,1:].values,df1_train.iloc[:,-1].values)
```

```python
model2.fit(df_train.iloc[:,1:].values,df2_train.iloc[:,-1].values)
```

```python
model3.fit(df_train.iloc[:,1:].values,df3_train.iloc[:,-1].values)
```

```python
model4.fit(df_train.iloc[:,1:].values,df4_train.iloc[:,-1].values)
```

```python
predict1=model1.predict(df_test.iloc[:,1:].values)
predict2=model2.predict(df_test.iloc[:,1:].values)
predict3=model3.predict(df_test.iloc[:,1:].values)
predict4=model4.predict(df_test.iloc[:,1:].values)
```

```python
pd1=pd.concat([pd.DataFrame(df_test.iloc[:,0].values,columns=['id']),pd.DataFrame(['A' for i in range(14000)],columns=['Seq_tag']),pd.DataFrame(predict1,columns=['Predict'])],axis=1)
pd2=pd.concat([pd.DataFrame(df_test.iloc[:,0].values,columns=['id']),pd.DataFrame(['B' for i in range(14000)],columns=['Seq_tag']),pd.DataFrame(predict2,columns=['Predict'])],axis=1)
pd3=pd.concat([pd.DataFrame(df_test.iloc[:,0].values,columns=['id']),pd.DataFrame(['C' for i in range(14000)],columns=['Seq_tag']),pd.DataFrame(predict3,columns=['Predict'])],axis=1)
pd4=pd.concat([pd.DataFrame(df_test.iloc[:,0].values,columns=['id']),pd.DataFrame(['D' for i in range(14000)],columns=['Seq_tag']),pd.DataFrame(predict4,columns=['Predict'])],axis=1)
```

```python
pd_out=pd.concat([pd1,pd2,pd3,pd4])
```
